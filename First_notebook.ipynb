# 1 create simple table using python on spark

```python
# From a list
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, schema=columns)
df.show()
```

### some operations on crateted DF(dataframe)

```python
# Select columns
df.select("Name").show()

# Filter rows
df.filter(df.Age > 30).show()

# Add a new column
df = df.withColumn("Age_plus_5", df.Age + 5)
df.show()

# Group by
df.groupBy("Age").count().show()
```

### To run same as SQL query on DF 

```python
# Create a temp view
df.createOrReplaceTempView("people")

# Run SQL
spark.sql("SELECT Name, Age FROM people WHERE Age > 30").show()
```

```python
from pyspark.sql.functions import col

# Example: Filter and select
df.filter(col("Age") > 25).select("Name").show()
```

```python
df.display()
```
